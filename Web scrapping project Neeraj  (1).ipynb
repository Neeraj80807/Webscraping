```python
from bs4 import BeautifulSoup 
import requests


```

```python
try:
    source = requests.get('https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting')
    source.raise_for_status()
    
    soup = BeautifulSoup(source.text,'html.parser')
    print(soup)
    
except exception as e:
    print(e)
```

```python
player_name = soup.find_all('a' , class_ = 'text-hvr-underline text-bold cb-font-16')
player_name
```

```python
name = []
for i in range(0,len(player_name)):
    name.append(player_name[i].get_text())
name
```

```python
rating_ = soup.find_all('div' , class_ = 'cb-col cb-col-17 cb-rank-tbl pull-right')
rating_
```

```python
rating = []
for i in range(0,len(rating_)):
    rating.append(rating_[i].get_text())
rating
```

```python
country_ = soup.find_all('div' , class_ = 'cb-font-12 text-gray')
country_
```

```python
country = []
for i in range(0 , len(country_)):
    country.append(country_[i].get_text())
country
```

```python
import pandas as pd
```

```python
sp = pd.DataFrame()
```

```python
sp['Player'] = name
sp['Country'] = country
sp['Rating'] = rating
```
